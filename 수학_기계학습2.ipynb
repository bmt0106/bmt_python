# [STEP3] 데이터 시각화하기

```python
# 선수들이 선호하는 발의 종류 막대그래프 그리기

import matplotlib.pyplot as plt

fifa2019['Preferred Foot'].value_counts().plot(kind='bar')

plt.legend()
plt.show()
```
